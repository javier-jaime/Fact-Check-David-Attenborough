## Initial exploration of the Dataset

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Week-5-Visualization/world-development-indicators/Indicators.csv')
print(data.shape)
data.head()
```

### Indicator 1: Fertility Rate Total (births per woman)

```python
Fertility_Rate = data[data.IndicatorCode == 'SP.DYN.TFRT.IN']
Fertility_Rate_by_Year = Fertility_Rate.groupby(['Year']).mean().round(decimals=2)
Fertility_Rate_by_Year.transpose()
```

```python
plt.plot(Fertility_Rate_by_Year.index, Fertility_Rate_by_Year['Value'].values)
```

```python
# Fertility Rate Low Income Countries (births per woman)
Fertility_Rate_Low_Income = Fertility_Rate[Fertility_Rate['CountryName'] == 'Low income'].groupby(['Year']).mean().round(decimals=2)
Fertility_Rate_Low_Income.transpose()
```

```python
# Fertility Rate Middle Income Countries (births per woman)
Fertility_Rate_Mid_Income = Fertility_Rate[Fertility_Rate['CountryName'] == 'Middle income'].groupby(['Year']).mean().round(decimals=2)
Fertility_Rate_Mid_Income.transpose()
```

```python
# Fertility Rate High Income Countries (births per woman)
Fertility_Rate_High_Income = Fertility_Rate[Fertility_Rate['CountryName'] == 'High income'].groupby(['Year']).mean().round(decimals=2)
Fertility_Rate_High_Income.transpose()
```

```python
# Fertility by income levels (births per woman)
plt.plot(Fertility_Rate_Low_Income.index, Fertility_Rate_Low_Income['Value'].values, 'r-', label='Low Income')
plt.plot(Fertility_Rate_Mid_Income.index, Fertility_Rate_Mid_Income['Value'].values, 'y-', label='Mid Income')
plt.plot(Fertility_Rate_High_Income.index, Fertility_Rate_High_Income['Value'].values, 'b-', label='High Income')
# Labels
plt.xlabel('Year')
plt.ylabel('births per woman')
plt.title('Fertility rate by income levels')
plt.legend()
plt.show()
```
